```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,6))
ax1=fig.add_subplot(2,2,1)
ax1.plot([1,2,3,4],[1,4,9,16])
ax1.set_title("Subplot 1(line)")

ax2 = fig.add_subplot(2,2,2)
ax2.scatter([1,2,3,4],[1,2,3,4], color = 'red')
ax2.set_title("Subplot 1(scatter)")

ax3 = fig.add_subplot(2,2,3)
ax3.bar(['A','B','C','D'], [10,20,30,40], color ='green')
ax3.set_title("Subplot 1(bar)")

ax4 = fig.add_subplot(2,2,4)
ax4.hist([3,4,5,6,6,6,7,8,9], bins = 5, color = 'purple')
ax4.set_title("Subplot 1(Histogram)")

plt.show()
```

```python
fig,axs = plt.subplots(2,1,sharex=True,figsize=(6,4))
axs[0].plot([1,2,3,4],[1,4,9,16], marker='o', color = 'blue')
axs[0].set_title('Sub-plot 1(Line)')
axs[0].set_ylabel('Y axis for sub-plot 1')
axs[1].scatter([1,2,3,4],[0,3,2,4], marker = 'x', color = 'red')
axs[0].set_title('Sub-plot 2 (Scatter)')
axs[0].set_ylabel('Y axis for sub-plot 2')

plt.tight_layout()
plt.show()
```

```python
x = [1,2,3,4,5]
y = [10,15,13,17,20]

fig, axs = plt.subplots(figsize=(8,4))

axs.plot(x,y,label='Line Plot', marker='o',linestyle='-')
axs.bar(x,y,width=0.3,label='Bar Plot', align = 'center', alpha = 0.6, color = 'green')

axs.scatter(x,y,label='Scatter Plot', color = 'red', marker='x', s=50 )

axs.set_title('Different chart types on the same axes')
axs.set_xlabel('X-Axis')
axs.set_ylabel('Y-Axis')

axs.legend()
plt.show()
```

```python
x1 = [1,2,3,4,5]
y1 = [10,15,13,17,20]

x2 = [1,2,3,4,5]
y2 = [5,8,16,2,15]

fig, ax = plt.subplots(figsize=(8,4))

ax.plot(x1,y1,label = 'Line Chart 1', marker = 'o', linestyle = '-', color = 'blue')
ax.plot(x2,y2,label = 'Line Chart 2', marker = 's', linestyle = '--', color = 'red')

ax.set_title("Two line charts in the same Plot")
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')

ax.legend()
plt.show()
```
